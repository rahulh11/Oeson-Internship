```python
#Libraries required in the assignment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
```

## Task 1: Data Generation

```python
#Number of samples required
N = 1000

#setting required mean and standard deviation for age, weight, height and income
#then creating 1000 samples of each category 
age = np.random.normal(35, 10, N)
height = np.random.normal(170, 15, N)
weight = np.random.normal(70, 10, N)
income = np.random.normal(50000, 15000, N)

#creating 1000 samples for gender with 50% probability each for male and female. 
gender = np.random.choice(['Male', 'Female'], size = N, p = [0.5, 0.5])

#create the data frame 
data = pd.DataFrame({'Age': age, 
             'Height': height,
              'Weight': weight,
             'Gender': gender,
              'Income': income})

#creates a data file with all the information.
data.to_csv('Synthetic_Datasheet_Assignment1.csv')

#print the first 10 rows of data 
print(data.head(5))
```

## Task 2: Descriptive Statistics

```python
#print the mean value of mean, age, height, weight and income
print("Mean")
print(data.mean(),"\n")

#print the median value of mean, age, height, weight and income
print("Median")
print(data.median(), "\n")

#print the variance of mean, age, height, weight and income
print("Variance")
print(data.var(),"\n")

#print the standard deviation of mean, age, height, weight and income
print("Standard Deviation")
print(data.std(), "\n")

#print the mode of gender 
print("Mode")
print(data['Gender'].value_counts())
```

## Task 3: Data Visualisations
### Histograms

```python
#Create subplot 4 histograms: Age, weight, height and income 

fig, axs = plt.subplots(2,2, figsize = (12,10))

#Age Histogram
axs[0,0].hist(data['Age'], bins = 25, color = 'red')
axs[0,0].set_title("Age Histogram")

#Height Histogram 
axs[0,1].hist(data['Height'], bins = 25, color = 'blue')
axs[0,1].set_title("Height Histogram")

#Weight Histogram
axs[1,0].hist(data['Weight'], bins = 25, color = 'green')
axs[1,0].set_title("Weight Histogram")

#Income Histogram
axs[1,1].hist(data['Income'], bins = 25, color = 'orange')
axs[1,1].set_title("Income Histogram")
```

### Histograms with KDE

```python
#Create subplot of all 4 histograms with KDE
fig, axs = plt.subplots(2, 2, figsize = (12,10))

#Age Histogram with KDE
sns.histplot(data['Age'], kde = True, ax = axs[0,0], color = 'red')
axs[0,0].set_title("Age Histogram with KDE")

#Height histogram with KDE
sns.histplot(data['Height'], kde = True, ax = axs[0,1], color = 'blue')
axs[0,1].set_title("Height Histogram with KDE")

#Weight Histogram with KDE
sns.histplot(data['Weight'], kde = True, ax = axs[1,0], color = 'orange')
axs[1,0].set_title("Weight Histogram with KDE")

#Income histogram with KDE
sns.histplot(data['Income'], kde = True, ax = axs[1,1], color = 'green')
axs[1,1].set_title("Income Histogram with KDE")
```

### Boxplots

```python
#Create subplot of 4 boxplots with age, height, weight and income. 
fig, axs = plt.subplots(2, 2, figsize = (12,10))

#Boxplot of Age
sns.boxplot(x = 'Age', data = data, ax = axs[0,0], color = 'red')
axs[0,0].set_title("Boxplot: Age")

#Boxplot of Height 
sns.boxplot(x = 'Height', data = data, ax = axs[0,1], color = 'blue')
axs[0,1].set_title("Boxplot: Height")

#Boxplot of Weight 
sns.boxplot(x = 'Weight', data = data, ax = axs[1,0], color = 'orange')
axs[1,0].set_title("Boxplot: Weight")

#Boxplot of Income
sns.boxplot(x = 'Income', data = data, ax = axs[1,1], color = 'yellow')
axs[1,1].set_title("Boxplot: Income")
```

## Task 4: Correlation Analysis

```python
#create a correlation matrix between age, height, weight and income
#this with calculate the correlation coefficient between all the different variables
#and create a matrix of values with them
corr_matrix = data.corr()

#plot the correlation matrix with heat map indicating the correlation level. 
#if the grid is red, it has a high correlation 
#if the grid is blue, it has little to no correlation. 
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot= True, cmap = 'coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

## Task 5: Inferential Statistics: 

```python
#Null Hypothesis: if p_value>0.05, we assume that there is no relationship between male and female incomes. 
#Alternate Hypothesis: if p_value<0.05, we reject null hypothesis. 

#generate male and female incomes
male_income = data[data['Gender'] == 'Male']['Income']
female_income = data[data['Gender'] == 'Female']['Income']

#calculate the p_value between male and female income. 
t_statistic, p_value = ttest_ind(male_income, female_income)
print("p value is:", p_value)


alpha = 0.05
#determine if we accept or rejct null hypothesis.
if p_value < alpha: 
    print("We reject the null hypothesis")
else: 
    print("We accept the null hypothesis")
```
