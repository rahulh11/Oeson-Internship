## Basic Plotting with Matplotlib 
### Line Chart

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
import plotly.subplots as sp
```

```python
#Question: Create a line plot for the function y = sin(x)

#Answer

#Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

#Create the Plot 
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('Simple Line Plot')
plt.grid('on')
plt.show() 
```

### Scatter Plot

```python
#Question: Create a scatter plot using random values/data

#Answer

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('Simple Scatter Plot')
plt.grid('on')
plt.show() 
```

### Bar Chart


```python
#Question: create a bar plot on the basis of given categories and values 

#Answer
categories = ['A', 'B', 'C']
values = [10, 25, 32]

plt.bar(categories,values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title('Simple Bar Plot')
plt.show() 
```

```python
#Question: Create a line plot for the function y = sin(x) and y = cos(x) w legends

#Answer

#Generate data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

#Create the Plot 
plt.plot(x, y1, label = 'sin(x)')
plt.plot(x, y2, label = 'cos(x)')
plt.xlabel("X-ais")
plt.ylabel("Y-axis")
plt.title('Line Plot with legend')
plt.grid('on')
plt.show() 
```

### Subplot

```python
#Question: Create a subplot for the function y = sin(x) and y = cos(x)

#Answer

#Generate data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

#Create the Plot 
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(x, y1)
ax1.set_title('sin(x)')
ax2.plot(x, y2)
ax2.set_title('cos(x)')
plt.tight_layout #ensures that all parameters on each plot is correct
plt.show()
```

```python
#1: Line PLot: Create a line plot of the functions y = x and y =x2 on the same graph. Add a legend

x = np.linspace(-10, 10, 100)
y1 = x
y2 = x*x

plt.plot(x,y1, label = 'y=x')
plt.plot(x,y2, label = 'y=x^2')
plt.legend()
plt.show()
```

```python
#2: Scatter plot: Create  a scatter plot using randfom data for x and y. Colour the points based on their value

x = np.random.rand(100)
y = np.random.rand(100)

# Create a scatter plot
plt.scatter(x, y, c=y, cmap='viridis')

# Add color bar
plt.colorbar(label = 'Colour Values')

# Add labels and title
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Scatter Plot of Random Data')

# Show the plot
plt.show()
```

```python
#3: Subplots: Create a 2x2 grid of subplots. Each subplot should contain a diff plot type with random data 

x = np.linspace(0, 10, 100)
y1 = np.sin(x)

y2 = np.exp(x)

categories = ['A', 'B', 'C']
values = [10, 25, 32]

x2 = np.random.normal(170, 10, 250)
x1 = np.random.rand(100)
y3 = np.random.rand(100)

#Create the Plot 
fig, axs = plt.subplots(2,2, figsize=(12,10))
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('sin(x)')
axs[1, 0].bar(categories, values)
axs[1, 0].set_title('Bar graph')
axs[0, 1].hist(x2)
axs[0, 1].set_title('Histogram')
axs[1, 1].scatter(x1, y3)
axs[1, 1].set_title('Scatter plot')
plt.show()
```

```python
# Example data: average monthly temperatures (in degrees Celsius) and standard deviations
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
average_temperatures = [5, 6, 9, 13, 17, 20, 22, 21, 18, 14, 9, 6]
std_devs = [1, 1.5, 2, 1.7, 2.5, 2, 1.8, 1.6, 1.9, 2.1, 1.4, 1.2]

# Create a bar plot with error bars
plt.figure(figsize=(12, 6))
plt.bar(months, average_temperatures, yerr=std_devs, capsize=5, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Average Temperature (C)')
plt.title('Average Monthly Temperature with Standard Deviation')

# Show the plot
plt.show()
```

## Plotly 


```python
#Creating interactive plots 
x = np.linspace(0,10,100)
y = np.sin(x)
fig = go.Figure(data=go.Scatter(x=x,y=y, mode='lines', name='sin(x)'))
fig.update_layout(title='Simple Line Plot', xaxis_title = 'X-axis', yaxis_title = 'Y-axis')
fig.show()
```

```python
#Question: Create an interactive scatter plot using random data

#Answer

x = np.random.rand(100)
y = np.random.rand(100)
fig = go.Figure(data=go.Scatter(x=x, y=y, mode = 'markers', name='Random Scatter'))
fig.update_layout(title='Simple Scatter Plot', xaxis_title = 'X-axis', yaxis_title = 'Y-axis')
fig.show()
```

```python
#Question: Create an interactive bar plot for the given categories and values

#Answer
categories = ['A', 'B', 'C']
values = [10, 20, 15]

fig = go.Figure(data=go.Bar(x=categories, y=values))
fig.update_layout(title='Simple Bar Plot', xaxis_title = 'Categories', yaxis_title = 'Valyes')
```

```python
#Question: Create subplots for y = sin(x) and y = cos(x) using Plotly

#Answer

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig = sp.make_subplots(rows=2, cols=1)
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='sin(x)'), row=1, col=1)
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='cos(x)'), row=2, col=1)
fig.update_layout(title = 'Subplot of sin and cos')
fig.show()
```
